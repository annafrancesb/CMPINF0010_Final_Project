```python
import pandas as pd
import geopandas as gpd
%matplotlib inline
playingFields = pd.read_csv("https://data.wprdc.org/datastore/dump/6af89346-b971-41d5-af09-49cfdb4dfe23")

byLat = playingFields
del byLat["has_lights"]
del byLat["field_usage"]
del byLat["goal_post"]
del byLat["infield_type"]
del byLat["left_field_distance"]
del byLat["right_field_distance"]
del byLat["center_field_distance"]
del byLat["backstop"]
del byLat["council_district"]
del byLat["ward"]
del byLat["tract"]
del byLat["public_works_division"]
del byLat["pli_division"]
del byLat["police_zone"]
del byLat["fire_zone"]
byLat.head()
#byLat.title('A Map Of Pittsburgh\'s Open Fields', fontsize=20)
byLat.plot.scatter(figsize=(8,8), x='latitude', y='longitude')
```

# Interpreting The Chart
